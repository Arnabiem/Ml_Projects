```python
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(obj):
  L=[]
  count=0
  for i in ast.literal_eval(obj):
    if count!=3:
      L.append(i['name'])
      count+=1
    else:
       break 
  return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head(2)
```

```python
def fetch_dirctor(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew']=movies['crew'].apply(fetch_dirctor)
```

Tag Stemming

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  l=[]
  for i in text.split():
    l.append(ps.stem(i))
  return " ".join(l)    
```

```python
new_df['Tags']=new_df['Tags'].apply(stem)
new_df['Tags'][0]
```

```python
ps.stem('impression')
```

Vectorization of tags

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
# stop words are articles,prepositions,conjunctions etc....
```

```python
vectors=cv.fit_transform(new_df['Tags']).toarray()
```

```python
vectors[0].shape
```

```python
cv.get_feature_names()
```

```python
new_df
```

cosine distance finding

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
similarity.shape
```

```python
similarity[3]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
new_df[new_df['title']=="The Dark Knight Rises"].index[0]
```

Recommendation Function 


```python
def recommend(movie):
  movie_index=new_df[new_df['title']==movie].index[0]
  distances=similarity[movie_index]
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movies_list:
    print(new_df.iloc[i[0]].title)  
```

```python
recommend('Avatar')
```
